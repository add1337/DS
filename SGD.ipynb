# Написать класс, который ищет минимальное значение при помощи стохастического градиентного спуска

```python
import numpy as np
import matplotlib.pyplot as plt
import random
# Найти минимум функции sqrt(x)
# class 
class StochasticGradientDesent(object):
    def __init__(self):
        pass
    @staticmethod
    # Функция
    def function(x):
        return np.sqrt(x)
    @staticmethod
    # Производная функции
    def function_derivative(x):
        return 1/(2*np.sqrt(x))
    
    # Алгоритм градиентного спуска
    def gradient_descent(self, n_iter:int, learning_rate: float, eps: float):
        
        solutions, scores, iters, error = list(), list(), list(), list()
        
  
        
        for i in range(n_iter):
            x_next = np.array([random.uniform(1,10)])
            x_pred = x_next.copy()
            
            gradient = self.function_derivative(x_next)
            
            x_next = x_next - learning_rate * gradient
     
            gradient_eval = self.function(x_next)
        
            solutions.append(x_next)
            scores.append(gradient_eval)
            error.append(abs(np.mean(x_next - x_pred)))
            iters.append(i)
            if  abs(np.mean(x_next - x_pred)) < eps:
                break

        return solutions, scores, iters, error
```

```python
# При помощи созданного класса найти минимальное значение функции learning_rate=0.01, eps=0.05
grd = StochasticGradientDesent()
solutions, scores, iters, error = grd.gradient_descent(n_iter=100, learning_rate=0.01, eps=0.05)
print(min(scores))
```
